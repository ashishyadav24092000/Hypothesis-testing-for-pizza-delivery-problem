# QUESTION
## 1. The mean response times for a random sample of 30 pizza deliveries is 32 minutes.
## 2. The population standard deviation is belived to be 10 minutes.
## 3. The pizza delivery services director wants to perform hypothesis test,with α =0.05(i.e. significance level = 5%), to determine whether the service goal of 30 minutes or less is being achieved?

# SOLUTION:


## ----------------------------------Data given to us::--------------------------------------------------------

## SAMPLE:                                                        
        x̄ = sample mean = 32 min
        n = sample size = 30
       
## POPULATION:
        α = significance level = 0.05
        µ = population mean = 30 min
        σ = population standard deviation = 10 min


## METHOD -1 ----> Hypothesis testing using p-value approach:
    
    Step-1:  Develop the hypotheses.
        
            Null hypothesis ----> H0 : µ ≤ 30(pizza delivery services is meeting the response goal time)
            Alternative hypothesis -----> Ha : µ > 30(pizza delivery services is not meeting the response goal time)
    
    
    Step-2: Specify the Level of Significance
        
             α = significance level = 0.05
    
    
    Step-3: Compute the value of test statistics.
             
             z = (x-μ)/(σ/√n) = (32-30)/(10/√30) = 1.09
            





## So z = 1.09

## Finding the area value at z = 1.09 

    Step-4: Compute the p-value for the calculated z

```python
        ## this will give the area-value at that calculated z-point or p-value
        from scipy.stats import norm 
        p_value  = 1-norm.cdf(1.09)
        print("For the calculated z, The value of p = ",p_value)
```

    Step-5: Determine whether to reject H0

## SINCE, 
        p-value > α
    i.e. 0.1378 > 0.05
##  this implies ---->  we do not reject H0.


# FINAL INSIGHT OR INFERENCE = There are not much statistical  evidence to infer that pizza delivery services is not meeting the response goal time.  

# METHOD-2 ------> Hypothesis testing using critical value approach:

    Step-1:  Develop the hypotheses.
        
            Null hypothesis ----> H0 : µ ≤ 30(pizza delivery services is meeting the response goal time)
            Alternative hypothesis -----> Ha : µ > 30(pizza delivery services is not meeting the response goal time)
    
    
    Step-2: Specify the Level of Significance
        
             α = significance level = 0.05
     
     
     Step-3: Compute the value of test statistics.
             
             z = (x-μ)/(σ/√n) = (32-30)/(10/√30) = 1.09
     
     
     
## So z = 1.09

    Step-4:  Determine the critical value and rejection rule 
            
            
            for α = 0.05 -----> Z_α = 1.645           // Z-α value obtained from standard z-distribution table
            
            
        NOW: 
            Reject H0 if Z  ≥ Z_α otherwise do not reject H0        //Critical value approach
            
            
            
    Step-5: Determine whether to reject H0
    
    
            Because,  test statistic Z = 1.09  and Z_α = 1.645
     this implies---> Z < Z_α i.e Z is not ≥ Z_α.
     
     
##    So we do not Reject H0

# FINAL INSIGHT OR STATISTICAL INFERENCE ::: There are not sufficient statistical evidence to infer that pizza delivery service is not meeting the response goal of 30 minutes.
